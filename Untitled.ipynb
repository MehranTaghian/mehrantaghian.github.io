```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
def create_wordcloud(text):
    # Create WordCloud object
#     wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)
    wordcloud = WordCloud(width=1000, height=400, background_color='white', max_words=50).generate(text)

    # Display the generated word cloud using matplotlib
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    wordcloud.to_file("docs/public/skills.jpg")
    plt.show()
```

```python
import yaml
yaml_file_path = 'skills.yml'

# Open the YAML file and load its content into a Python dictionary
with open(yaml_file_path, 'r') as file:
    data = yaml.safe_load(file)

data
```

```python
text = ''
for k, v in data.items():
    text += ' ' * v  + k + ' ' * v 
```

```python
create_wordcloud(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Your skills and their respective emphasis
skills_emphasis = {
    'Python': 5,
    'Machine Learning': 4,
    'Data Analysis': 3,
    'JavaScript': 2,
    'Web Development': 3,
    'Communication': 4,
    'Problem Solving': 5,
    'Teamwork': 3,
}

# Generate a string based on skill emphasis
text = ' '.join([skill + ' ' * emphasis for skill, emphasis in skills_emphasis.items()])

# Generate a word cloud with different emphasis
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
